```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

dfs = pd.read_csv('instance/excel/titanic_data.csv')
default_age = dfs['Age'].isnull()
age = dfs['Age'].values
age
```

```python
age = np.reshape(age, (-1, 1))
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp.fit(age)
dfs['Age'] = imp.transform(age)
```

```python
to_array = []
for ag in age:
    lst = []
    for num in ag:
        num += 100
        lst.append(num)
    to_array.append(lst)

to_array
```

```python
replacements = []
range_age = len(age)
range_age = [x for x in range(1, range_age, 2)]

for ag in age[[range_age]]:
    lst = []
    for num in ag:
        num += 100
        lst.append(num)
    replacements.append(lst)

for idx, replacement in zip(range_age, replacements):
    age[idx] = replacement

age
```
