```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

dfs = pd.read_csv('instance/excel/titanic_data.csv')
default_age = dfs['Age'].isnull()
age = dfs['Age'].values
age
```

```python
age = np.reshape(age, (-1, 1))
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp.fit(age)
dfs['Age'] = imp.transform(age)
```

```python
to_array = []
for ag in age:
    lst = []
    for num in ag:
        num += 100
        lst.append(num)
    to_array.append(lst)

to_array
```

```python
new_elements = []
replacements = []
range_age = len(age)
range_age = [x for x in range(1, range_age, 2)]

for ag in age[[range_age]]:
    lst = []
    for num in ag:
        num += 100
        lst.append(num)
    replacements.append(lst)

for idx, replacement in zip(range_age, replacements):
    age[idx] = replacement
```

```python
range_age = len(age)
for idx in range(1, range_age, 2):
    age[idx] += 100

age
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn import datasets, tree

x = dfs[['Pclass', 'Fare']]
y = dfs[['Survived']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.5)
model = KNeighborsClassifier()
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
acc = accuracy_score(y_test, y_predict)
acc
```

```python
wine = datasets.load_wine()
x = wine.data
y = wine.target
my_classifier = LogisticRegression()
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.5)
my_classifier.fit(x_train, y_train)
prediction = my_classifier.predict(x_test)
acc = accuracy_score(y_test, prediction)
acc

```
