```python
import cv2 as cv
import numpy as np
import pandas as pd
from google.cloud import vision
import io
import re
import os

os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = r'config/config_ocr.json'

client = vision.ImageAnnotatorClient()
image_path = 'instance/uploads/user_pic.png'
with io.open(image_path, 'rb') as image_file:
    content = image_file.read()
```

```python
image = vision.Image(content=content)
response = client.document_text_detection(image=image)
document = response.full_text_annotation
```

```python
pages = document.pages
for page in pages:
    for block in page.blocks:
        bounding_box = [(vertices.x, vertices.y) for vertices in block.bounding_box.vertices]
    for block in page.blocks:
        for paragraphs in block.paragraphs:
            print('paragraphs confidence', paragraphs.confidence)
            for words in paragraphs.words:
                word_text = ''.join([symbol.text for symbol in words.symbols])
                print(word_text)
                for symbol in words.symbols:
                    print('symbol: {} (confidence: {})'. format(symbol.text, symbol.confidence))

```
